# INTRODUCCION

## Funcion Seno y Parametos 

```python
import numpy as np
import matplotlib.pyplot as plt

def mi_funcion_sen(vmax, dc, ff, ph, nn, fs):
    ts = 1/fs
    tt = np.arange(nn) * ts
    
    xx = vmax * np.sin(2 * np.pi * ff * tt + ph) + dc
    
    return tt, xx

# --- Parámetros para ver 1 Hz ---
fs = 100        # Muestreamos 100 veces por segundo (suficiente para 1 Hz)
N = 100         # Tomamos 100 muestras (así tenemos exactamente 1 segundo)
vmax = 1        # 1 Volt de amplitud
dc = 0          # Sin offset
f1 = 1          # frecuencia de la senal 1 Hz
fase = 0        # 0 radianes
```

  En la estructura de la funcion seno primero empezamos por definir el tiempo entre muestras (ts) es el recíproco de la frecuencia de muestreo del ADC, luego generamos un vector de "tiempo" llamado tt que al multiplicarlo por ts, se transforman esos índices en valores de tiempo reales en segundos y por ultimo aplica la fórmula matemática del seno a todo el vector de tiempo simultáneamente para calcular el valor de voltaje correspondiente a cada muestra de la senal.

Los parametros se definen por:

1. vmax (Amplitud): Es el valor máximo de tensión (voltios) que alcanza la señal desde el centro hasta su pico.

2. dc (Valor medio): Es el desplazamiento vertical de la señal; define el nivel de corriente continua sobre el que oscila el seno.

3. ff (Frecuencia): Indica cuántos ciclos completos realiza la señal en un segundo, medido en Hertz (Hz).

4. ph (Fase): Es el desplazamiento horizontal inicial de la onda en el tiempo cero, medido en radianes.

5. nn (Número de muestras): Es la cantidad total de puntos digitales que el ADC tomará para representar la señal.

6. fs (Frecuencia de muestreo): Es la velocidad del ADC, es decir, cuántas muestras captura por cada segundo (muestras/s).

## Teorema de Nyquist y Efecto Aliasing

  Para que una señal analógica se convierta a digital sin perder su identidad, debemos seguir las reglas del "tráfico de datos".
  
  1. Frecuencia de Nyquist ($f_N$)
    Es la frecuencia máxima que un sistema de muestreo puede capturar correctamente. Se define como la mitad de la frecuencia de muestreo ($f_s$):$$f_N = \frac{f_s}{2}$$Regla de oro: Si querés medir una señal de frecuencia $f$, tu frecuencia de muestreo debe ser al menos el doble ($f_s > 2f$). Si no cumplís esto, la información se destruye.

  3. Aliasing (Solapamiento)
    El Aliasing es un error de interpretación digital. Ocurre cuando muestreamos una señal demasiado lento (por debajo del límite de Nyquist).Como el ADC no tiene suficientes puntos para "dibujar" la onda rápida, los une de forma incorrecta, creando una señal falsa de baja frecuencia (un "alias") que no existe en la realidad.

## IMPLEMENTACION FUNCION 

## Grafico funcion seno de 1 Hz

```python
# Definimos la funcion de 1Hz
tt, xx = mi_funcion_sen(vmax=vmax, dc=dc, ff=f1, ph=fase, nn=N, fs=fs)

# Graficamos f1 = 1 Hz
plt.figure(figsize=(10, 4))
plt.plot(tt, xx, 'b-o', markersize=3, label=f'Senoidal {f1} Hz')
plt.title(f"Señal Senoidal de {f1} Hz (1 ciclo completo)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.grid(True)
plt.legend()
plt.show()
```

Visualizamos el ciclo completo de una onda sinusoidal de 1 Hz (periodo 1 segundo) porque la duración total del tiempo de observación ($T_{total} = nn / fs$) coincide exactamente con el tiempo que le toma a la señal completar una oscilación. Al configurar los parámetros para que el producto de la cantidad de muestras por el intervalo entre ellas sea igual al periodo de la función ($1/ff$), el gráfico captura el recorrido desde el inicio ($0^\circ$) hasta el final de la fase ($360^\circ$).
